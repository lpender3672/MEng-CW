```python
import numpy as np 
import matplotlib.pyplot as plt

plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "serif"
plt.rcParams['figure.facecolor'] = 'white'
```

```python
I0	= 0.56028
IG	= 0.084
C	= 0.02090554102
I1  = 0.016
J   = 0.025
M	= 10.8
a	= 0.21
g = 9.81
omega3 = 4650 * 2 * np.pi / 60
```

```python
precession_vs_theta = np.loadtxt("precession_vs_theta.txt")
```

```python
# plot

theta = np.linspace(80 , 160 , 100) * np.pi/180
Q_mg = a * np.sin(theta) * 2 * g
plt.plot(theta * 180/np.pi, Q_mg, label = r'$Q_{mg}$', linestyle='--')

plt.scatter(precession_vs_theta[:,0], precession_vs_theta[:,3], label = r'$Q_{\Omega}$', marker='x', color = 'r')

plt.ylabel(r'$Q$ [Nm]')
plt.xlabel(r'$\theta$ [$^o$]')
plt.legend( loc='lower left', prop={'size': 12})
plt.grid()
# set ticks
plt.xticks(np.arange(90, 151, 15))

plt.savefig("precession_vs_theta.png", dpi=300)
plt.show()
```

```python

# load nutation data
nutation = np.loadtxt('nutation_data.txt')
print(nutation.shape)
```

```python
# plot
plt.scatter(nutation[:,1], nutation[:,2], label = r'$f_{meas}$', marker='x', color = 'r')

f = C * omega3 / IG * (1 + J/IG * (1 / np.tan(theta)**2) + I1/IG * (1 / np.sin(theta) ** 2)) ** -0.5
plt.plot(theta * 180 / np.pi, f / (2 * np.pi), label = r'$f_{(A7)}$', linestyle='--')
print(f / (2 * np.pi))

plt.xlabel(r'$\theta$ [$^o$]')
plt.ylabel(r'Frequency [Hz]')
plt.legend( loc='lower left', prop={'size': 12})
plt.grid()
# set ticks
plt.xticks(np.arange(90, 151, 15))
plt.ylim(0, 20)

plt.savefig("nutation.png", dpi=300)
plt.show()
```
