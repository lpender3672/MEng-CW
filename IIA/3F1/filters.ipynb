```python
import scipy as sp
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

# butterworth filter

```python
# plot the frequency response of a butterworth filter

butter = sp.signal.butter(4, 10, 'low', analog=True)
bessel = sp.signal.bessel(4, 10, 'low', analog=True)
cheby1 = sp.signal.cheby1(4, 5, 10, 'low', analog=True)
cheby2 = sp.signal.cheby2(4, 5, 10, 'low', analog=True)
ellip = sp.signal.ellip(4, 5, 40, 10, 'low', analog=True)

def plot_filter(filter, name = ""):
    w, h = sp.signal.freqs(filter[0], filter[1])
    plt.semilogx(w, 20 * np.log(abs(h)), label=name)

plot_filter(butter, "Butterworth")
plot_filter(bessel,  "Bessel")
plot_filter(cheby1,  "Chebyshev Type I")
plot_filter(cheby2,  "Chebyshev Type II")
plot_filter(ellip,   "Elliptic")

plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.grid(which='both', axis='both')
plt.legend(loc='best')
plt.savefig('filter.png',dpi=300)
plt.show()
```
