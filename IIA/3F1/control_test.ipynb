```python
import numpy as np
from matplotlib import pyplot as plt

import control as ct
```

```python

sys = ct.tf((1), (1, 0.2, 2, 0))

t = np.linspace(0, 10, 1000)

t, y = ct.step_response(sys, t)

plt.plot(t, y)
plt.show()
```

```python
# bode and nyquist plot
ct.bode_plot(sys)
plt.show()

ct.nyquist_plot(sys)
plt.xlim(-3, 1)
plt.ylim(-2, 2)
plt.show()

ct.nichols_plot(sys)
plt.show()
```

## Discrete time system

```python

dt = 0.01

sys = ct.tf((1), (1, 0.2, 2, 0), dt=dt)

t = np.arange(0, 8, dt)

t, y = ct.step_response(sys, t)

plt.plot(t, y)
plt.show()
```

```python
## plots

ct.bode_plot(sys)
plt.show()
```
