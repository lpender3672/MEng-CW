```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

theta = 10 * np.linspace(- np.pi, np.pi, 1000)
z = np.exp(1j * theta)
A = 1 / (1 - z**-1)

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(theta, np.abs(A))
plt.ylim(0, 100)
plt.title('Magnitude')
plt.subplot(2, 1, 2)
plt.plot(theta, np.angle(A))
plt.title('Phase')

plt.show()
```
