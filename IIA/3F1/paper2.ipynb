```python
import numpy as np
import control as ct
import matplotlib.pyplot as plt
```

# Q1

```python
# discrete time system of z
z = ct.TransferFunction.z
G1 = 1 / (z**2 * (z - 1))
G2 = (4*z - 2) / (3 * (z - 1)**2)
G3 =  4 / (z - 1)**3
```

```python
# nyquist plots

# G1
ct.nyquist_plot(G1)
plt.ylim(-3, 3)
plt.xlim(-3, 1)
```

```python

ct.nyquist_plot(G2)
plt.xlim(-3, 1)
plt.ylim(-3, 3)
```

```python

ct.nyquist_plot(G3)
plt.xlim(-5, 20)
plt.ylim(-5, 20)
```
