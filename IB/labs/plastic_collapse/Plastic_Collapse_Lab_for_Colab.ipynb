### **Plan A: Hands-on lab instructions**

You will carry out the instructions in this section during your lab session.

### **Simply Supported Beam**
1. Check that the span of the simply supported beam that is set up in the rig is 200 mm.
2. One student should draw up a table in their lab book with 2 columns, headings “Force (N)” and “Deflection (mm)”, or enter the data directly into the Jupyter Google Sheet. Label the table as “Simply supported beam results table”.
3. The other student should check the loading hooks are secure on the middle of the beam. After this check is completed, they can now start turning the circular nut under the rig (beneath the load cell) to load the beam.
4. The larger display at the top of the rig measures the load in N (using the loading cell), the smaller display close to the beam measures the displacement in mm (using the displacement transducer).
5. Record a data point every 10N until the load no longer increases; this should be at about 60N. Do not stop loading once the load reaches a peak; we are interested in the behaviour of the simply supported beam as it collapses.
6. Continue loading and now record data points every 2mm, until about 15mm.
7. Make sure the structure has visibly deformed and there is a kink (plastic hinge) at the midpoint of the simply supported beam.
8. Take a few data points as you unload the beam; observe the permanently deformed shape.










simplysupported.JPG

###**Portal frame test**

1. Remove the simply supported beam and supports from the rig and slide over the portal frame that is set up towards the loading hook.
2. One student should draw up a table in their lab book with 2 columns, headings “Force (N)” and “Deflection (mm)” or enter the data directly into the Jupyter Google Sheet. Label the table as “Portal frame results table”.
3. The other student should secure the loading hooks on to both the loading beam and the portal frame. Check with the demonstrator if in doubt. After this check is completed, they can now start turning the circular nut under the rig (beneath the load cell) to load the portal frame.
4. As before, the larger display at the top of the rig measures the load in N (using the loading cell), the smaller display close to the portal frame measures the displacement in mm (using the displacement transducer).
5. Record a data point every 25N until a load of 150N. Then continue in increments of 10N until the load no longer increases; this should be at about 200–250N.
6. Do not stop loading once the load reaches a peak; we are interested in the behaviour of the portal frame as it collapses.
7. Continue loading and now record data points every 1 mm, until about 13 mm. Make sure the structure has visibly deformed and there is a kink (plastic hinge) at the midpoint of the horizontal part of the portal frame.
8. Take a few data points as you unload the portal frame; observe the permanently deformed shape.
9. One student takes a photo of the collapsed portal frame (after fully unloading).

apparatus.png


###**Swapping from portal frame to propped cantilever**

1. One student should now loosen the loading nut to fully unload the frame. Unhook the loading hooks from the loading beam and portal frame, then remove the loading beam.
2. Now remove the portal frame from the rig. Turn the large screw underneath the base clamps to release the clamps and remove the portal with the base clamps from the rig.
3. Set up the single clamped support and the roller support as in the photo below for the propped cantilever test, and place them 300mm apart, ensuring the loading cell is centrally between them.
4. Loosen the clamped support and get the mild steel rod. Clamp one end of the rod in the clamped support; use the full length of the clamp to secure as much of the beam as possible – otherwise it may behave more like a simply supported beam. Check the span of the beam is 300mm (from the top of the roller to where the rod exits the clamp).
5. Fasten the loading cell into position and hook the loading hook over the rod at the centre of the span.
6. Place the displacement transducer end on the rod right next to the loading hook, while also ensuring the transducer is as vertical as possible.
7. Check with the demonstrator that your setup is correct before continuing.

###**Propped cantilever test**

1. As before, the larger display at the top of the rig measures the load in N (using the loading cell), the smaller display just above the beam measures the displacement in mm (using the displacement transducer).
2. One student should draw up another table in their lab book with headings “Force (N)” and “Deflection (mm)”  or enter the data directly into the Jupyter Google Sheet. Label the table as “Propped cantilever results table”.
3. The beam can now be loaded, but this time record data points every 10N until the load no longer increases; this should be at approximately 50–80N.
4. As before, continue loading past this point, but now record data points every 1mm, until about 12mm. As the displacements get larger, you may need to re-adjust the displacement transducer. Make sure that a kink (plastic hinge) has formed in the centre of the beam.
5. Take a few data points as you unload the propped cantilever; observe the permanently deformed shape.
6. One student takes a photo of the collapsed propped cantilever (after fully unloading).
7. Please ensure you can contact the other member(s) of your lab group after the lab, as you’ll need to share your results tables and photos of collapsed structures within your pair or group of three in order to complete the post-lab report.

```python
worksheet = gc.open('PlasticCollapse').worksheet('Simplysupportedbeam') #Rename your Google Sheet here

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()

# Convert to a DataFrame.
df1 = pd.DataFrame.from_records(rows)

worksheet = gc.open('PlasticCollapse').worksheet('Portalframe') #Rename your Google Sheet here

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()

# Convert to a DataFrame.
df2 = pd.DataFrame.from_records(rows)

worksheet = gc.open('PlasticCollapse').worksheet('Proppedcantilever') #Rename your Google Sheet here

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()

# Convert to a DataFrame.
df3 = pd.DataFrame.from_records(rows)

#Tidy up the data.
df1 = df1.drop([0])
df1 = df1.rename(columns={0:'Force (N)', 1:'Deflection (mm)'})

df2 = df2.drop([0])
df2 = df2.rename(columns={0:'Force (N)', 1:'Deflection (mm)'})

df3 = df3.drop([0])
df3 = df3.rename(columns={0:'Force (N)', 1:'Deflection (mm)'})

#Render the dataframe.

print("Simply supported beam")
display(df1)
print("Portal frame")
display(df2)
print("Propped cantilever beam")
display(df3)
```

This looks great, but there's an issue which is not immediately obvious. If you have experience with Python you might have already spotted it.

When we imported the data from the Google Sheet using `get_all_values()`, this imported the data as a "string" datatype. This datatype is useful for things like text, but if we try to plot graphs with numbers saved as strings, this causes issues. This is easily solved by converting the datatype of our `DataFrame` from a "string" to a "float" (floating point number, which is just a decimal number), which is what the code cell below does.

You'll notice the table looks the same, but see what happens later if you skip out this code cell! This a common issue in Python, and it's worth remembering to check your datatypes if you're having unexpected problems in future projects!

```python
df1 = df1.astype('float')
df2 = df2.astype('float')
df3 = df3.astype('float')

print("Simply supported beam")
display(df1)
print("Portal frame")
display(df2)
print("Propped cantilever beam")
display(df3)
```

### Plot graphs

The `matplotlib.pyplot` object is used to plot graphs of results of Force (N) vs. Deflection (mm).



```python
plt.figure(figsize=(26,8))

plt.subplot(1, 3, 1) 
plt.plot(df1['Deflection (mm)'], df1['Force (N)'])
plt.scatter(df1['Deflection (mm)'], df1['Force (N)'])
plt.title("Simply Supported Bean")
plt.xlabel('Deflection (mm)')
plt.ylabel('Force (N)')
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.8, ls='-', lw=1)

plt.subplot(1, 3, 2) 
plt.plot(df2['Deflection (mm)'], df2['Force (N)'])
plt.scatter(df2['Deflection (mm)'], df2['Force (N)'])
plt.title("Portal Frame")
plt.xlabel('Deflection (mm)')
plt.ylabel('Force (N)')
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.8, ls='-', lw=1)

plt.subplot(1, 3, 3) 
plt.plot(df3['Deflection (mm)'], df3['Force (N)'])
plt.scatter(df3['Deflection (mm)'], df3['Force (N)'])
plt.title("Propped Cantilever Beam")
plt.xlabel('Deflection (mm)')
plt.ylabel('Force (N)')
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.8, ls='-', lw=1)

plt.show()
```
