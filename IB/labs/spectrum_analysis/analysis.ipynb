```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "serif"
```

```python
f_triangle = [
0.998,
2.992,
4.986,
6.98,
8.987]

v_triangle = [
546.4,
62.9,
22.28,
10.88,
4.895
]

f_square = [
0.998,
2.998,
4.995,
6.992,
8.991
]

v_square = [
846.1,
258,
159.6,
98.89,
73.18
]

v_triangle = np.array(v_triangle)
v_triangle = v_triangle/v_triangle[0]

v_square = np.array(v_square)
v_square = v_square/v_square[0]
```

```python

f = np.linspace(0, 10, 1000)

# plot as points
plt.plot(f_square, v_square, 'o', label='Square', color='blue')

# fit 1/n to square
def square_fit(n, a, b):
    return a / n + b

popt, pcov = curve_fit(square_fit, f_square, v_square)
asq, bsq = np.round(popt, 2).astype(str)
label_sq = 'Square fit: $'+ asq +'n^{-1}'+bsq+'$'
plt.plot(f, square_fit(f, *popt), label=label_sq, color='blue')

plt.plot(f_triangle, v_triangle, 'o', label='Triangle', color='red')


# fit 1/n^2 to triangle
def triangle_fit(n, a, b):
    return a / n**2 + b

popt, pcov = curve_fit(triangle_fit, f_triangle, v_triangle)
at, bt = np.round(popt, 2).astype(str)
label_t = 'Triangle fit: $'+ at +'n^{-2} + '+bt+'$'
plt.plot(f, triangle_fit(f, *popt), label=label_t, color='red')



plt.xlim(0, 10)
plt.ylim(0, 1.2)
plt.xlabel('n')
plt.ylabel('Relative Fourier Coefficient')
plt.grid()
plt.legend()

plt.savefig('square_triangle.png', dpi=300)
```
