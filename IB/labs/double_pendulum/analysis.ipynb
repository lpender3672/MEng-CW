```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "serif"
```

```python
# read csv

df = pd.read_csv('exercises1234.csv')

t1 = np.array(df.iloc[0])[1:].astype(float)
a1 = np.array(df.iloc[1])[1:].astype(float)
b1 = np.array(df.iloc[2])[1:].astype(float)
plt.plot(t1, a1, 'b-', label=r'$\alpha$')
plt.plot(t1, b1, 'r-', label=r'$\beta$')

plt.xlabel('Time [s]')
plt.ylabel('Angle [deg]')
plt.title('Mode 1 at 1.0575 Hz')
plt.legend()
```
