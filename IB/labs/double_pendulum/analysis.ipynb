```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["text.usetex"] = True
plt.rcParams["font.family"] = "serif"

%matplotlib inline
```

```python
# read csv

df = pd.read_csv('exercises1234.csv')

t1 = np.array(df.iloc[0])[1:].astype(float) - 1
a1 = np.array(df.iloc[1])[1:].astype(float)
b1 = np.array(df.iloc[2])[1:].astype(float)
plt.plot(t1, a1, 'b-', label=r'$\alpha$')
plt.plot(t1, b1, 'r-', label=r'$\beta$')

plt.xlabel('Time [s]')
plt.ylabel('Angle [deg]')
plt.title('Mode 1 at 1.0575 Hz')
plt.legend()
```

```python
# subplots of modes

fig, axs = plt.subplots(2, 1, sharex=True)
axs[0].plot(t1, a1, 'b-', label=r'$\alpha$')
axs[0].plot(t1, b1, 'r-', label=r'$\beta$')
axs[0].set_title('Mode 1 at 1.0575 Hz')
axs[0].legend()

axs[1].plot(t2, a2, 'b-', label=r'$\alpha$')
axs[1].plot(t2, b2, 'r-', label=r'$\beta$')
axs[1].set_title('Mode 2 at 0.1910 Hz')
#axs[1].legend()

plt.xlabel('Time [s]')
fig.text(0.04, 0.5, 'Angle [Deg]', va='center', rotation='vertical')
# sup label
plt.suptitle('Double pendulum modes')
plt.tight_layout()

plt.savefig('modes.png', dpi=300)

plt.show()
```

```python
# animate

%matplotlib tk
import matplotlib.animation as animation


a1 = a5
b1 = b5

fig, ax = plt.subplots()

ax.set_xlim(( -3, 3))
ax.set_ylim((-3, 3))

line, = ax.plot([], [], 'o-', lw=2)

def init():
    line.set_data([], [])
    return (line,)

def animate(i):
    x = [0, np.sin(a1[i]*np.pi/180), np.sin(a1[i]*np.pi/180) + np.sin(b1[i]*np.pi/180)]
    y = [0, -np.cos(a1[i]*np.pi/180), -np.cos(a1[i]*np.pi/180) - np.cos(b1[i]*np.pi/180)]
    line.set_data(x, y)
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=a1.size, interval=20, blit=True)

# save
#anim.save('double_pendulum.mp4', fps=30, extra_args=['-vcodec', 'libx264'])

plt.show()

#%matplotlib inline
```
