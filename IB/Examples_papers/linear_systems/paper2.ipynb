```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.facecolor'] = 'white'
plt.tight_layout()
```

```python
def G(s):
    return 22000 / (s+22000)

# draw bode plot
w = np.logspace(0, 8, 1000)
s = 1j*w
mag = 20 * np.log10(np.abs(G(s)))
phase = np.angle(G(s), deg=True)
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.semilogx(w, mag)
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Magnitude [dB]')
plt.grid()
plt.subplot(122)
plt.semilogx(w, phase)
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Phase [deg]')
plt.grid()
plt.show()

```

```python

def G(s):
    return 22000*s / ((s+300)*(s+22000))

# draw bode plot
w = np.logspace(0, 3, 1000)
s = 1j*w
mag = 20 * np.log10(np.abs(G(s)))
phase = np.angle(G(s), deg=True)
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.semilogx(w, mag)
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Magnitude [dB]')
plt.grid()
plt.subplot(122)
plt.semilogx(w, phase)
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Phase [deg]')
plt.grid()
plt.show()

```
