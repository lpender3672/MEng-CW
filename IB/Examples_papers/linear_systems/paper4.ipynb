```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import control as ctr
%matplotlib inline
plt.rcParams['figure.facecolor'] = 'white'
plt.tight_layout()
```

```python
s = ctr.tf('s')
# Step response
H,m = 1121, 200
kp, Td = 0.091, 2.0
tau = 1e3
delay = 1/(1+s*tau)
#L = kp* (1 + s*Td) * np.exp(- tau * s) *H / (m*s**2) # not possible for some reason
L = kp* (1 + s*Td) * delay * H / (m*s**2)
G = L/(1+L)
# draw nyquist plot
ctr.nyquist(L)
plt.show()
```
